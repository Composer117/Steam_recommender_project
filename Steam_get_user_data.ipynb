```python
import pandas as pd

game_data = pd.DataFrame(sample_data["response"]["games"])
game_data = game_data[game_data["playtime_forever"] != 0]
game_data = game_data[["appid", "name", 'playtime_forever']]
game_data = game_data.sort_values("playtime_forever", ascending=False)
game_data["playtime_forever"] = game_data["playtime_forever"]/60
game_data
```
