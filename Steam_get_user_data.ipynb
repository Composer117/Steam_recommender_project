```python
from steam.steamid import steam64_from_url


def get_user_game_data(steam_url, steam_games_data = steam_games):
    steamID = steam64_from_url(steam_url)
    sample_data = api.call('IPlayerService.GetOwnedGames', steamid=steamID, include_appinfo=True, include_played_free_games=True, appids_filter=True, include_free_sub=True, language="EN", include_extended_appinfo=True)
    game_data = pd.DataFrame(sample_data["response"]["games"])
    game_data = game_data[game_data["playtime_forever"] != 0]
    game_data = game_data[["appid", "name", 'playtime_forever']]
    game_data = game_data.sort_values("playtime_forever", ascending=False)
    game_data["playtime_forever"] = game_data["playtime_forever"]/60
    user_game_data = game_data.merge(steam_games_data, how="inner", left_on="appid", right_on="App ID")
    user_game_data.drop(["appid", "name"], axis=1, inplace=True, errors='ignore')
    return user_game_data
```

```python
def get_friend_list_df(steam_url):
    apicall = api.ISteamUser.GetFriendList(steamid = steam64_from_url(steam_url), relationship="all")
    friend_list = apicall["friendslist"]["friends"]
    friend_list = pd.DataFrame(friend_list)
    friend_list = friend_list["steamid"].tolist()
    friend_list_string = ", ".join(friend_list)
    friend_df = pd.DataFrame(api.ISteamUser.GetPlayerSummaries(steamids=friend_list_string)["response"]["players"])
    return friend_df
```
